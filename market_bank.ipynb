```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score
import os
import pickle
data=pd.read_csv("bank-full.csv",delimiter=";")
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data=data.drop(["contact", "poutcome"],axis=1)
```

```python
data
```

```python
en=LabelEncoder()
features=['job','marital','default','housing','loan','month','y']
for i in features:
    data[i]=en.fit_transform(data[i].astype("str"))
```

```python
data["education"]=data['education'].fillna("unknown")
encoder=['unknown','primary','secondary','tertiary']
for i, word in enumerate(encoder):
    data['education']=data['education'].astype("str").str.replace(word,str(i))
```

```python
data['education']=data['education'].astype('int64')
```

```python
data.info()
```

```python
outliers={}
for i in range(data.shape[1]):
    min_t=data[data.columns[i]].mean()-(3*data[data.columns[i]].std())
    max_t=data[data.columns[i]].mean()+(3*data[data.columns[i]].std())
    count=0
    for j in data[data.columns[i]]:
        if j<min_t or j>max_t:
            count+=1
            outliers[data.columns[i]]=[count,data.shape[0]]
outliers       
```

```python
X=data.drop('y',axis=1)
Y=data["y"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
dev=X_test.shape[0]/X_train.shape[0]
```

```python
X_train,X_dev,Y_train,Y_dev=train_test_split(X_train,Y_train,test_size=dev,random_state=2)
```

```python
print(X_train.shape,X_dev.shape,X_test.shape,Y_train.shape,Y_dev.shape,Y_test.shape)
```

```python
model_DC=tree.DecisionTreeClassifier(random_state=2)
model_DC=model_DC.fit(X_train,Y_train)
```

```python
model_e1=MLPClassifier(random_state=2)
model_e1=model_e1.fit(X_train,Y_train)
```

```python
sets=["training",'validation','testing']
X_set=[X_train,X_dev,X_test]
Y_set=[Y_train,Y_dev,Y_test]
```

```python
ouptut={"DC":{"acc":[],"pre":[]},"MLP":{"acc":[],"pre":[]}}
for i in range(0,len(sets)):
    Y_pred1=model_DC.predict(X_set[i])
    ouptut["DC"]["acc"].append(accuracy_score(Y_set[i],Y_pred1))
    ouptut["DC"]["pre"].append(precision_score(Y_set[i],Y_pred1))
    Y_pred2=model_e1.predict(X_set[i])
    ouptut["MLP"]["acc"].append(accuracy_score(Y_set[i],Y_pred2))
    ouptut["MLP"]["pre"].append(precision_score(Y_set[i],Y_pred2))
    
```

```python
ouptut
```

```python
model_tree =tree.DecisionTreeClassifier(random_state = 2, min_samples_leaf=100, 
                                    max_depth=100)
model_tree.fit(X_train, Y_train)
```

```python
model_NN = MLPClassifier(random_state = 2, max_iter=1000,
                         hidden_layer_sizes = [100,100,50,25,25], tol=1e-4)
model_NN.fit(X_train, Y_train)
```

```python
ouptut1={"DC":{"acc":[],"pre":[]},"MLP":{"acc":[],"pre":[]}}
for i in range(0,len(sets)):
    Y_pred3=model_tree.predict(X_set[i])
    ouptut1["DC"]["acc"].append(accuracy_score(Y_set[i],Y_pred3))
    ouptut1["DC"]["pre"].append(precision_score(Y_set[i],Y_pred3))
    Y_pred4=model_NN.predict(X_set[i])
    ouptut1["MLP"]["acc"].append(accuracy_score(Y_set[i],Y_pred4))
    ouptut1["MLP"]["pre"].append(precision_score(Y_set[i],Y_pred4))
    
```
