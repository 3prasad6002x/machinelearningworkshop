```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score,precision_score,recall_score
import numpy as np
```

```python
data=load_breast_cancer()
```

```python
X=pd.DataFrame(data.data)
Y=pd.DataFrame(data.target)
```

```python
X_new,X_test,Y_new,Y_test=train_test_split(X,Y,test_size=0.1,random_state=101)
```

```python
test_size=X_test.shape[0]/X_new.shape[0]
```

```python
X_train,X_val,Y_train,Y_dev=train_test_split(X_new,Y_new,test_size=test_size,random_state=101)
```
